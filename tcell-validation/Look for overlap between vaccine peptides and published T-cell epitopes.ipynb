```python
df_human_infection = df[(df.Species == "Human") & (df.Exposure == "Infection")]
```

```python
df_vaccine_peptides = pd.read_csv("../vaccine-peptides/final-vaccine-peptides-27mer.csv")
```

```python
df_vaccine_peptides
```

```python
k = 9
```

```python
from collections import defaultdict
vaccine_peptide_to_kmers = defaultdict(set)

for p in df_vaccine_peptides.Sequence:
    for i in range(len(p) - k + 1):
        kmer = p[i:i+k]
        vaccine_peptide_to_kmers[p].add(kmer)
        
```

```python
vaccine_peptide_to_kmers
```

```python
tcell_epitopes_to_kmers = defaultdict(set)
for p in df_human_infection.Peptide:
    for i in range(len(p) - k + 1):
        kmer = p[i:i+k]
        tcell_epitopes_to_kmers[p].add(kmer)
        
```

```python
tcell_epitopes_to_kmers
```

```python
vaccine_peptides_to_overlapping_tcell_epitopes = defaultdict(set)
for v_p, v_ks in vaccine_peptide_to_kmers.items():
    for t_p, t_ks in tcell_epitopes_to_kmers.items():
        common_kmers = v_ks.intersection(t_ks)
        if len(common_kmers) > 0:
            vaccine_peptides_to_overlapping_tcell_epitopes[v_p].add(t_p)
            
```

```python
len(vaccine_peptides_to_overlapping_tcell_epitopes)
```

```python
vaccine_peptides_to_overlapping_tcell_epitopes
```

```python
vaccine_peptides_to_sources = defaultdict(set)
for v_p, t_ps in vaccine_peptides_to_overlapping_tcell_epitopes.items():
    for t_p in t_ps:
        vaccine_peptides_to_sources[v_p].update(
            df_human_infection[df_human_infection.Peptide == t_p].Source)
```

```python
vaccine_peptides_to_num_sources = {v_p: len(sources) for (v_p, sources) in vaccine_peptides_to_sources.items()}
```

```python
vaccine_peptides_to_num_sources
```

```python
from collections import Counter
vaccine_peptides_to_n_responding = Counter()
vaccine_peptides_to_n_total = Counter()
for v_p, t_ps in vaccine_peptides_to_overlapping_tcell_epitopes.items():
    for t_p in t_ps:
        df_sub = df_human_infection[df_human_infection.Peptide == t_p]
        for n_r, n_t in zip(df_sub["# Responding"], df_sub["# Total"]):
            if np.isnan(n_r) or np.isnan(n_t):
                continue
            vaccine_peptides_to_n_responding[v_p] += n_r
            vaccine_peptides_to_n_total[v_p] += n_t
```

```python
vaccine_peptides_to_n_responding
```

```python
vaccine_peptides_to_n_total
```

```python
vaccine_peptides_to_frac_responding = {
    v_p: n_r / vaccine_peptides_to_n_total[v_p] 
    for (v_p, n_r) in vaccine_peptides_to_n_responding.items()
}
```
